```python
model = keras.models.Sequential([
    Flatten(input_shape = (28,28)),
    Dense(128,    activation = "relu"),
    Dropout(0.2),
    Dense(10, activation = "softmax")
    
])
```

```python
model.compile(
    optimizer = "adam",
    loss = "sparse_categorical_crossentropy",
    metrics = ["accuracy"]
)
```

```python
def plot_image(i, predict_array, y_test1, x_test1):
    predict_array , y_test1, x_test1 = predict_array[i], y_test1[i], x_test1[i]
    
    
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x_test1, cmap = plt.cm.binary)
    
    predict_label = np.argmax(predict_array)
    
    if predict_label == y_test1:
        color = "blue"
    else:
        color = "red"
    plt.xlabel("{} {:2.0f}% ({})".format(  class_names[predict_label], 
                                           100*np.max(predict_array), 
                                           class_names[y_test1],
                                           color = color)      )
    
def plot_value_array(i, predict_array, y_test2):
    predict_array, y_test2 = predict_array[i] , y_test2[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    
    thisplot = plt.bar(range(10) , predict_array, color = "#777777")
    plt.ylim([0,1])
    predict_label = np.argmax(predict_array)
    thisplot[predict_label].set_color("red")
    thisplot[y_test2].set_color("blue")
    
    
```
