### Spam Detection Using Machine Learning

+ Use our ML Model as a package

```python
# Load EDA Pkgs
import pandas as pd 
import numpy as np
```

```python
# Load ML Pkgs
# ML Packages For Vectorization of Text For Feature Extraction
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
```

```python
# Visualization Packages
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset
df = pd.read_csv("data/spamdataset.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df[['content','class']].head()
```

```python
df1 = df[['content','class']]
```

```python
# Extract Feature With CountVectorizer
corpus = df['content']
cv = CountVectorizer()
X = cv.fit_transform(corpus) # Fit the Data
```

```python
cv.get_feature_names()
```

```python
spam_vectorizer_model = open("spam_vectorizer.pkl","wb")
joblib.dump(cv,spam_vectorizer_model)
```

```python
spam_vectorizer_model.close()
```

```python
y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(df.shape)
print(X_train.shape)
```

```python
# Naive Bayes Classifier

nb = MultinomialNB()
nb.fit(X_train,y_train)
```

```python
# Accuracy Score
nb.score(X_test,y_test)
```

```python
# Sample Prediciton of A Text if it is a spam
comment = ["Check this out"]
vect = cv.transform(comment).toarray()
```

```python
#class_dict = {'ham':0,'spam':1}
nb.predict(vect)
```

```python
# Sample Prediciton 2
comment1 = ["Great song Friend"]
vect2 = cv.transform(comment1).toarray()
nb.predict(vect2)
```

```python
# Save Models
```

```python
import joblib
```

```python
spam_detector_nb_model = open("spam_detector_nb_model.pkl","wb")
joblib.dump(nb,spam_detector_nb_model)
```

```python
spam_detector_nb_model.close()
```

### Using Logistic Regression

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
spam_detector_rf_model = open("spam_detector_rf_model.pkl","wb")
joblib.dump(rf,spam_detector_rf_model)
spam_detector_rf_model.close()
```
